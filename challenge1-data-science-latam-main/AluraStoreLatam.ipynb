### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#Lista de tiendas
precio = [tienda['Precio'], tienda2['Precio'], tienda3['Precio'], tienda4['Precio']]
print(precio)
```

```python
#Suma de ventas de tiendas
lista_ventas_tienda = []
i = 1
print('---- Ventas totales por tienda ----')
for tiendax in precio:

  venta_tienda = list(tiendax)
  lista_ventas_tienda.append(sum(venta_tienda))
  print(f'\n---- Tienda {i} ----')
  i+=1
  print(f'Ventas totales: {lista_ventas_tienda[-1]}')
```

```python
#grafico de ventas de tienda
import matplotlib.pyplot as plt
tiendas_grafica = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
colores = ['blue', 'green', 'red', 'purple'] # Define colors for each bar

plt.bar(tiendas_grafica, lista_ventas_tienda,color=colores)
plt.xlabel('Tiendas')
plt.ylabel('Ventas')
plt.title('Ventas por tienda')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


def plot_category_pie(df, title):
    category_sales = df['Categoría del Producto'].value_counts()
    plt.figure(figsize=(4, 4))
    plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', startangle=90)
    plt.title(title)
    plt.axis('equal')
    plt.show()


plot_category_pie(tienda, 'Ventas por Categoría - Tienda 1')
plot_category_pie(tienda2, 'Ventas por Categoría - Tienda 2')
plot_category_pie(tienda3, 'Ventas por Categoría - Tienda 3')
plot_category_pie(tienda4, 'Ventas por Categoría - Tienda 4')

```

# 3. Calificación promedio de la tienda


```python
# Calcular la calificación promedio de clientes por tienda
def average_rating(df, store_number):
    average = df['Calificación'].mean()
    print(f"Calificación promedio de la tienda {store_number}: {average}")
    return average

average_rating(tienda, 1)
average_rating(tienda2, 2)
average_rating(tienda3, 3)
average_rating(tienda4, 4)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


store_numbers = [1, 2, 3, 4]
average_ratings = [
    average_rating(tienda, 1),
    average_rating(tienda2, 2),
    average_rating(tienda3, 3),
    average_rating(tienda4, 4),
]


plt.figure(figsize=(8, 6))
plt.bar(store_numbers, average_ratings, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.xlabel("Número de Tienda", fontsize=12)
plt.ylabel("Calificación Promedio", fontsize=12)
plt.title("Calificación Promedio de Clientes por Tienda", fontsize=14)
plt.xticks(store_numbers)
plt.grid(axis='y', linestyle='--', alpha=0.7)


for i, v in enumerate(average_ratings):
    plt.text(i + 1, v + 0.05, f"{v:.2f}", ha='center', fontsize=10)

plt.show()
```

# 4. Productos más y menos vendidos

```python
# prompt: En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def top_bottom_products(df, store_number):
    """
    Identifica y visualiza los productos más y menos vendidos en una tienda.

    Args:
        df (DataFrame): DataFrame con los datos de ventas
        store_number (int): Número de la tienda
    """
    # 1. Calcular ventas por producto
    ventas_por_producto = df.groupby('Producto')['Precio'].sum().sort_values(ascending=False) # Sort by the 'Precio' column after grouping and summing

    # 2. Identificar top y bottom productos
    top_products = ventas_por_producto.head(3)
    bottom_products = ventas_por_producto.tail(3)

    # 3. Mostrar resultados en consola
    print(f"\nTienda {store_number}:")
    print("Top 3 productos más vendidos:")
    print(top_products)
    print("\nTop 3 productos menos vendidos:")
    print(bottom_products)

    # 4. Visualización
    plt.figure(figsize=(12, 6))

    # Gráfico de los más vendidos
    plt.subplot(1, 2, 1)
    sns.barplot(x=top_products.values, y=top_products.index, palette='viridis')
    plt.title(f'Tienda {store_number} - Top 3 Productos')
    plt.xlabel('Cantidad Vendida')
    plt.ylabel('Producto')

    # Gráfico de los menos vendidos
    plt.subplot(1, 2, 2)
    sns.barplot(x=bottom_products.values, y=bottom_products.index, palette='magma')
    plt.title(f'Tienda {store_number} - Bottom 3 Productos')
    plt.xlabel('Cantidad Vendida')
    plt.ylabel('Producto')

    plt.tight_layout()
    plt.show()

# Cargar los datos (ya lo has hecho)
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Analizar cada tienda
top_bottom_products(tienda, 1)
top_bottom_products(tienda2, 2)
top_bottom_products(tienda3, 3)
top_bottom_products(tienda4, 4)
```

# 5. Envío promedio por tienda

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def calcular_costo_envio_promedio(df, store_number):
    """
    Calcula y visualiza el costo promedio de envío para una tienda.

    Args:
        df (DataFrame): DataFrame con los datos de ventas
        store_number (int): Número de la tienda
    Returns:
        float: Costo promedio de envío
    """
    # Calcular el promedio del costo de envío
    costo_promedio = df['Costo de envío'].mean()

    # Mostrar resultados
    print(f"\nTienda {store_number}:")
    print(f"Costo promedio de envío: ${costo_promedio:.2f}")

    return costo_promedio

# Cargar los datos
urls = [
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv",
    "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"
]

tiendas = [pd.read_csv(url) for url in urls]
resultados = []

# Calcular para cada tienda
for i, df_tienda in enumerate(tiendas, start=1):
    costo_prom = calcular_costo_envio_promedio(df_tienda, i)
    resultados.append({'Tienda': i, 'Costo_Promedio_Envio': costo_prom})

# Crear DataFrame con resultados
df_resultados = pd.DataFrame(resultados)

# Visualización comparativa
plt.figure(figsize=(10, 6))
sns.barplot(data=df_resultados, x='Tienda', y='Costo_Promedio_Envio', palette='coolwarm')
plt.title('Comparación del Costo Promedio de Envío por Tienda')
plt.ylabel('Costo Promedio ($)')
plt.xlabel('Número de Tienda')

# Añadir etiquetas con los valores
for index, row in df_resultados.iterrows():
    plt.text(row.name, row.Costo_Promedio_Envio, f'${row.Costo_Promedio_Envio:.2f}',
             color='black', ha="center")

plt.tight_layout()
plt.show()
```
